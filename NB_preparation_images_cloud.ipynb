# Analyse en Composantes principales sur un Ã©chantillon


### Import des librairies

````python
import pandas as pd
import os
from PIL import Image
````

### Mise en place du dataframe

````python
# liste des classes prÃ©sentes
classes = os.listdir("Training")

# dataframe vide
df = pd.DataFrame()

# traitement de chaque classe
for cl in classes:
    print(f"Traitement de {cl}")
    count = 0
    for file in os.listdir(f"Training/{cl}"):
        print(f"Image {count}")
        count += 1
        new_line = pd.Series(cl)
        new_line = new_line.append(pd.Series(list(Image.open(f"Training/{cl}/{file}").tobytes())), ignore_index=True)
        df = df.append(new_line, ignore_index=True)
````

````python
# enregistrement du dataframe

df.to_csv("three_cat.csv")
````

### Analyse en composantes principales

````python
from sklearn import decomposition

pca = decomposition.PCA(n_components=100)
pca.fit(df.drop(columns=[0]))

print("Pourcentage de variance expliquÃ©e : ")
print(pca.explained_variance_ratio_)
print("Pourcentage de variance expliquÃ©e cumulÃ©e :")
print(pca.explained_variance_ratio_.cumsum())
#print("Composantes principales : ")
#print(pca.components_)
#print("Valeurs propres : ")
#print(pca.explained_variance_)
````

````python
# visualisation des taux de variance expliquÃ©e

import matplotlib.pyplot as plt

fig = plt.figure()
plt.title("Eboulis des valeurs propres")
plt.plot(range(1,101),pca.explained_variance_ratio_, '-o')
plt.xlabel("valeurs propres")
plt.ylabel("Taux de variance expliquÃ©e")
plt.show

fig = plt.figure()
plt.title("Cumul Ã©boulis des valeurs propres")
plt.plot(range(1,101),pca.explained_variance_ratio_.cumsum(), '-o')
plt.xlabel("valeurs propres")
plt.ylabel("Taux de variance expliquÃ©e cumulÃ©e")
````

### Programme Pyspark

Ce programme rÃ©alise le prÃ©traitement complet :  
* Mise en place du dataframe
* Analyse en composantes principales
* Enregistrement des Ã©lÃ©ments de l'ACP et du dataframe final  

Cette version est celle utilisant les 4000 premiÃ¨res valeurs de chaque image. Elle est enregistrÃ©e dans le fichier *preparation_distant_4000.py*.

````python
from boto3 import client, resource
import os
from PIL import Image
from numpy import array
from pyspark.sql.functions import udf
from pyspark.sql.types import StringType
from pyspark.ml.linalg import Vectors
from pyspark.sql import SparkSession
from pyspark.ml.feature import PCA


print("DÃ©but de l'application")


# rÃ©cupÃ©ration de tous les noms de fichiers dans le dossier Training du bucket
s3 = resource("s3", region_name="us-east-2")
bucket = s3.Bucket("samuel-pate-projet-8")
list_files = [f.key[9:] for f in list(bucket.objects.all()) if f.key.startswith("Training")]

print("list_files remplie avec", len(list_files), "fichiers")

# mise en place d'un dictionnaire avec en clÃ© les classes et en valeurs la liste des fichiers de chacune d'elle
dict_classes = dict()
for f in list_files:
    classe = f[:f.index('/')]
    if classe in dict_classes.keys():
        dict_classes[classe].append(f)
    else:
        dict_classes[classe] = [f]

print("dict_classes rempli avec les clÃ©s :", dict_classes.keys())

# instanciation de SparkSession et SparkContext
spark = SparkSession.builder.appName("projet-8").getOrCreate()
print("spark instanciÃ©")
sc = spark.sparkContext
print("sc instanciÃ©")

# le dictionnaire des classes est rendu accessible Ã  toutes les machines du cluster
broad_classes = sc.broadcast(dict_classes)
print("dict_classes broadcastÃ©")



def fill_class(cl):
    """
    param : le nom d'une classe (string)

    return : la liste des couples (classe, liste des valeurs de pixels)
    pour chaque image de cette classe
    """
    s3 = resource("s3", region_name="us-east-2")
    bucket = s3.Bucket("samuel-pate-projet-8")
    result = list()
    for f in broad_classes.value[cl]:
        object = bucket.Object(f"Training/{f}")
        response = object.get()
        file_stream = response['Body']
        im = array(Image.open(file_stream))
        result.append((cl, Vectors.dense(im.astype("int").flatten().tolist()[:4000])))
    return result

# crÃ©ation du dataFrame contenant les couples (classe, liste de pixels) de toutes les images dans le dossier Training
print("RÃ©cupÃ©ration des images et mise en place du dataFrame")
rdd = sc.parallelize(dict_classes.keys())
rdd = rdd.flatMap(fill_class)
schema = ["label", "features"]
df_final = spark.createDataFrame(rdd, schema=schema)
print("DataFrame terminÃ©")

# Analyse en 50 composantes principales des listes de pixels
pca = PCA(k=50, inputCol="features", outputCol="pca_features")
pca.write().overwrite().save("s3a://samuel-pate-projet-8/Saved_from_cloud_4000/PCA_object")
print("Objet PCA instanciÃ© et sauvegardÃ©")

# entraÃ®nement de la PCA
model = pca.fit(df_final)
model.write().overwrite().save("s3a://samuel-pate-projet-8/Saved_from_cloud_4000/PCA_model")
print("PCA entraÃ®nÃ©e et modÃ¨le sauvegardÃ©")

# application de la PCA
df_final = model.transform(df_final).drop("features")
expl_var = model.explainedVariance.toArray()
print("Variance expliquÃ©e : ", expl_var)
print("Variance expliquÃ©e cumulÃ©e :", expl_var.cumsum())


def array_to_string(my_list):
    """
    param : my_list liste de valeurs

    return : la liste contenue dans my_list sous forme d'une chaÃ®ne de caractÃ¨re
    """
    return '[' + ','.join([str(elem) for elem in my_list]) + ']'

# enregistremement des features de la PCA 
# (sous forme de chaÃ®ne de caractÃ¨res pour pouvoir le mettre dans un fichier csv)
array_to_string_udf = udf(array_to_string, StringType())
df_final = df_final.withColumn("features_as_string", array_to_string_udf(df_final["pca_features"]))
df_final = df_final.drop("pca_features") 
print("Enregistrement du dataFrame")
df_final.write.mode("overwrite").csv("s3a://samuel-pate-projet-8/Saved_from_cloud_4000/dataframe_PCA_features")
````
